```python
import pandas as pd
import re
df = pd.read_csv('andic_dicts.csv', index_col=None, header=None)
id = []
segments = []
data = df[df[13].isin(['andi1255', 'tind1238', 'ghod1238', 'bagv1239', 'akhv1239', 'kara1474'])]
data_filtered = data.dropna(subset=[11])
data_filtered[7] = [str(i) for i in data_filtered[7]]
data_filtered_2 = data_filtered[data_filtered[7].isin(['nan'])]
data_filtered_2[7] = data_filtered_2[7].replace('nan', '')
good = data_filtered_2[[1, 2, 3, 4, 5, 6, 7, 8, 11, 12, 13, 14, 15]]
for i in good[5]:
    if str(i) == 'nan':
        i = ''
    segments.append(str(i).replace('-', ' '))
for i in range(1, len(good)+1):
    id.append(i)
good[3] = id
good['SEGMENTS'] = segments
good['ALIGNMENT'] = segments
final = good.rename({1: 'ID', 2: 'ID_MEAN', 3: 'ID_DICT', 4: 'LEMMA', 5: 'TRANSCRIPTION', 'SEGMENTS': 'SEGMENTS', 'ALIGNMENT': 'ALIGNMENT', 6: 'MORPHOLOGY', 7: 'BOR', 8: 'ANNOTATOR', 11: 'CONCEPT', 12: 'DEFINITION', 13: 'DOCULECT', 14: 'REFERENCE', 15: 'COMMENTS'}, axis=1)
final_2 = final[~final.CONCEPT.str.contains('см.')]
for_edictor = final_2[['ID', 'LEMMA', 'CONCEPT', 'DOCULECT', 'SEGMENTS', 'ALIGNMENT', 'TRANSCRIPTION']]
```

```python
cognates = pd.read_csv('cognates.csv', index_col=None, header=None)
cognates_2 = cognates[[0, 1, 2, 3, 4, 5]]
cognates_2.columns = ['COGNATE', 'akhv1239', 'bagv1239', 'ghod1238', 'kara1474', 'tind1238']
cognates_2['COGNATE'] = [str(i) for i in cognates_2['COGNATE']]
cognates_3 = cognates_2[~cognates_2['COGNATE'].isin(['nan'])]
cognates_3
```

```python
cognates_list_normal = []
cognates_list = cognates_3['COGNATE'].tolist()
for c in cognates_list:
    f = c.split('_')
    cognates_list_normal.append(f[0])
cog_id = []
for i in range(1, len(cognates_list)+1):
    cog_id.append(i)
```

```python
langs = ['akhv1239', 'bagv1239', 'ghod1238', 'kara1474', 'tind1238']
cognates_langs = cognates_3[['akhv1239', 'bagv1239', 'ghod1238', 'kara1474', 'tind1238']]
lis = cognates_langs.astype(str).values.tolist()
big = []
for l in lis:
    small = []
    for j, word in enumerate(l):
        if word != 'nan':
            small.append(langs[j])
        else:
            small.append('')
    big.append(small)
for k, word in enumerate(cog_id):
    big[k].append(word)
```

```python
cog_dict = {}
for i in range(len(cognates_list_normal)):
    cog_dict[cognates_list_normal[i]] = big[i]
```

```python
meanings_df = for_edictor['CONCEPT'].tolist()
langs_df = for_edictor['DOCULECT'].tolist()
```

```python
cog_id_df = []
for j, word in enumerate(meanings_df):
        if word in cog_dict.keys() and langs_df[j] in cog_dict[word]:
            cog_id_df.append(cog_dict[word][-1])
        else:
            cog_id_df.append('')
```
